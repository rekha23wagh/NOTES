```python
data4 = {'emp_id': [101,102,103,104],
        'salary' : [234,235,258,985],
         'pf': [23,24,25,26]
        }
```

```python
df4=pd.DataFrame(data4)
```

```python
df4
```

```python
data5 = {'emp_id' : [101,106,102,108],
        'mob_no' : [3435,3536,3637,3738],
        'house_no' : [1,2,3,4]
        }
```

```python
df5 = pd.DataFrame(data5)
```

```python
df5
```

```python
pd.merge(df4, df5)                      #merge operation takes common items only like emp_id and  101 and 102 only
```

```python
pd.merge(df4, df5, how ='left')    #left join i.e complte from left and common from right df4=left, df5=right in this case
```

```python
pd.merge(df4, df5, how ='right')  #right join i.e complte from right and common from left df4=left, df5=right in this case
```

```python
data6 = {'emp_id1': [101,102,103,104],
        'salary' : [234,235,258,985],
         'pf': [23,24,25,26]
        }

data7 = {'emp_id2' : [101,106,102,108],
        'mob_no' : [3435,3536,3637,3738],
        'house_no' : [1,2,3,4]
        }
```

```python
df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)
```

```python
df6
```

```python
df7
```

```python
pd.merge(df6, df7, left_on= 'emp_id1', right_on= 'emp_id2')   

                #even no column will same, on basis similar data its possible to merge
```
